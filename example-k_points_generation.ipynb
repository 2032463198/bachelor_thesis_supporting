示例代码1: 布里渊区离散坐标生成

在第一布里渊区最小对称部分生成均匀的离散动量坐标 (这个例子是二维的，形状满足C_4v对称性),存储在文件kpoints_original_x_this_example.txt和kpoints_original_x_this_example.txt中。

```python
import numpy as np

deltak0=1/40            #k-mesh fineness
a=2e-7                  #cell period along x direction
b=2e-7                   #cell period along y direction
filetitle="this_example"

kx=[]
ky=[]
Size=int(1/deltak0)
for i in range(0,Size+1):
    for j in range(i,Size+1):
        kx.append(j*deltak0*np.pi/a)
        ky.append(i*deltak0*np.pi/b)
np.savetxt('kpoints_original_x_%s.txt'%(filetitle), np.array(kx), newline=' ', fmt='%s')
np.savetxt('kpoints_original_y_%s.txt'%(filetitle), np.array(ky), newline=' ', fmt='%s')
np.savetxt('basic_info_%s.txt'%(filetitle), np.array([len(ky),deltak0,a,b]), newline=' ', fmt='%s')
print("Number of original k points: %d"%(len(kx)))
```
